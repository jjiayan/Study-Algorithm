### 6046 : [기초-비트시프트연산] 정수 1개 입력받아 2배 곱해 출력하기(설명)(py) 

```python
print(int(input()) * 2)
```

> 정수를 **2배로 곱하거나 나누어** 계산해 주는 **비트단위 시프트 연산자** `<<`, `>>`

| 연산 | 의미 | 예시 | 결과 |
|------|------|------|------|
| `<<` | 왼쪽 시프트 (×2) | `print(10 << 1)` | `20` |
| `>>` | 오른쪽 시프트 (÷2) | `print(10 >> 1)` | `5` |

```python
print(int(input()) << 1)
```

### 6083 : [기초-종합] 빛 섞어 색 만들기(설명)(py)

```python
r, g, b = map(int, input().split())
cnt = 0
for i in range(r):
    for j in range(g):
        for k in range(b):
            print(i, j, k)
            cnt += 1
print(cnt)
```

### 6096 : [기초-리스트] 바둑알 십자 뒤집기(py)

```python
arr = []
for _ in range(19):
    arr.append(list(map(int, input().split())))
n = int(input())

for _ in range(n):
    x, y = map(int, input().split())
    x -= 1
    y -= 1
    
    # 세로
    for i in range(19):
        arr[i][y] = arr[i][y] ^ 1
    # 가로
    for j in range(19):
        arr[x][j] = arr[x][j] ^ 1

for i in range(19):
    for j in range(19):
        print(arr[i][j], end=' ')
    print()
    
```

### 6097 : [기초-리스트] 설탕과자 뽑기(py)

```python
h, w = map(int, input().split())
arr = [[0] * w for _ in range(h)]
n = int(input())
for _ in range(n):
    l, d, x, y = map(int, input().split())
    x -= 1
    y -= 1
    # 0 가로 
    if d == 0:
        for j in range(y, y + l):
            arr[x][j] = 1
    # 1 세로
    else:
        for i in range(x, x + l):
            arr[i][y] = 1

for i in range(h):
    for j in range(w):
        print(arr[i][j], end=' ')
    print()
```

### 6098 : [기초-리스트] 성실한 개미(py)

```python
arr = []
for _ in range(10):
    arr.append(list(map(int, input().split())))
    
# (1, 1) 시작점 0
# 오른쪽으로 가다가 1만나면 아래로 이동 가능 [x+1, y], [x, y+1]
# 맨 아래 가장 오른쪽 도착시, 더 이상 움직일 수 없을 경우, 먹이를 찾은 경우 그곳에 머무름
x, y = 1, 1
arr[x][y] = 9

while (True):

    # 오른쪽 0
    if y + 1 < 10 and arr[x][y+1] == 0:
        arr[x][y+1] = 9
        y += 1
    # 오른쪽 2
    elif y + 1 < 10 and arr[x][y+1] == 2:
        arr[x][y+1] = 9
        break
    # 오른쪽 1
    else:
        # 아래 0
        if x + 1 < 10 and arr[x+1][y] == 0:
            arr[x+1][y] = 9
            x += 1
        # 아래 2
        elif x + 1 < 10 and arr[x+1][y] == 2:
            arr[x+1][y] = 9
            break
        # 아래 1
        else:
            break
    

for i in range(10):
    for j in range(10):
        print(arr[i][j], end=' ')
    print()
```
